# Data Visualization in Python

## LinePlot

#### Importing the libraries

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

### Loading the FMRI dataset

```python
fmr=pd.read_csv("fmri.csv")
fmr=sns.load_dataset("fmri")
```

### Printing the first five rows of data

```python
fmr.head()
```

```python
fmr.head(30)
```

```python
fmr.describe()
```

## Line Plot with Timepoint and Signal column data

```python
sns.lineplot(x="timepoint",y='signal',data=fmr)
```

## Mapping another column with a different color

```python

#sns.lineplot(x="timepoint",y='signal',data=fmr)
sns.lineplot(x="timepoint",y="signal",data=fmr,hue="event")
plt.show()
```

## Adding some styling for the event column data

```python
sns.lineplot(x="timepoint",y="signal",data=fmr,hue="event",style="event")
plt.show()
```

## Adding markers for value point 

```python
sns.lineplot(x="timepoint",y="signal",data=fmr,hue="event",style="event",marker=True)
plt.show()
```

## BarPlot 

#### importing Prokemon Dataset

```python
import pandas as pd
pokemon=pd.read_csv("pokemon.csv")
```

### Printing first 5 rows

```python
pokemon.head()
```

### Checking speed of legendary Pokemon

```python
sns.barplot(x="is_legendary",y="speed",data=pokemon)
plt.show()
```

### Checking weight of legendary Pokemon

```python
sns.barplot(x="is_legendary",y="weight_kg",data=pokemon)
plt.show()
```

### Adding a hue to show Pokemon generation

```python
sns.barplot(x="is_legendary",y="speed",data=pokemon,hue="generation")
plt.show()
```

## Using Palettes to assign different colour elements

```python
sns.barplot(x="is_legendary",y="speed",data=pokemon,palette="vlag")
plt.show()
```

```python
sns.barplot(x="is_legendary",y="speed",data=pokemon,color="red")
plt.show()
```

# Histogram / DistPlot

### Importing the Diamonds dataset

```python
diamonds=pd.read_csv("diamonds.csv")
```

## printng 5 rows

```python
diamonds.head()
```

### plotting the histogram based on price 

```python
sns.displot(diamonds['price'])
plt.show()
```

## Adding some colour to the distribution plot

```python
sns.displot(diamonds['price'],color="green")
plt.show()
```

# Ploting only the histogram without the frequency curve 

```python
sns.displot(diamonds['price'],kde=False,color="green") # kde=False  means Only print Histogram curve
plt.show()
```

## Changing the number of bins

```python
sns.displot(diamonds['price'],kde=False,color="green",bins=5)
plt.show()
```

# Box Plot

### importing dataset

```python
churn=pd.read_csv('churn.csv')
```

## printing first 5 rows

```python
churn.head()
```

## plotting based on churn and tenure columns

```python
sns.boxplot(x="Churn",y="tenure",data=churn)
plt.show()
```

```python
sns.boxplot(x="InternetService",y="MonthlyCharges",data=churn)
plt.show()
```

### Plotting based on the contact

```python
sns.boxplot(x="Contract",y="tenure",data=churn,palette="Set1")
plt.show()
```

## Changing line width of boxplot

```python
sns.boxplot(x="Contract",y="tenure",data=churn,palette="Set1",linewidth=5)
plt.show()
```

## Plotting  based on custom ordering on X axis

```python
sns.boxplot(x="Contract",y="tenure",data=churn,palette="Set1",linewidth=5, order=["One year","Two year","Month-to-month"],hue="PaymentMethod")
```
